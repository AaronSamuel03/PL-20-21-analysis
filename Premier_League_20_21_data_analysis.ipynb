```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#load dataset
pl_df = pd.read_csv('EPL_20_21.csv')
pl_df.head()
```

```python
pl_df.isnull()
```

```python
pl_df['MinsPerMatch'] = (pl_df['Mins']/pl_df['Matches']).astype(int)
pl_df['GoalsPerMatch'] = (pl_df['Goals']/pl_df['Matches']).astype(float)
```

```python
pl_df.head()
```

```python
Total_Goals = pl_df['Goals'].sum()
print(Total_Goals)
```

```python
Total_Penalty_Goals = pl_df["Penalty_Goals"].sum()
print(Total_Penalty_Goals)
```

```python
Total_Penalty_Attempts = pl_df['Penalty_Attempted'].sum()
print(Total_Penalty_Attempts)
```

```python
#Pie chart for penalties missed vs scored
plt.figure(figsize=(13,6))
pl_not_scored = pl_df['Penalty_Attempted'].sum() - Total_Penalty_Goals
data = [pl_not_scored, Total_Penalty_Goals]
labels = ['Penalties missed', 'Penalties scored']
color = sns.color_palette('Set2')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
#Positions
pl_df['Position'].unique()
```

```python
#All forwards
pl_df[pl_df['Position'] == 'FW']
```

```python
#Most players from which nation
nationality = pl_df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar', figsize=(12,6), color=sns.color_palette('magma'))
```

```python
#Players based on age group
Under20 = pl_df[pl_df['Age'] <= 20]
age20_25 = pl_df[(pl_df['Age'] > 20) & (pl_df['Age'] <= 25)]
age25_30 = pl_df[(pl_df['Age'] > 25) & (pl_df['Age'] <= 30)]
Above_30 = pl_df[pl_df['Age'] > 30]
```

```python
# Calculate the number of players in each age group
x = np.array([Under20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(), Above_30['Name'].count()])

# Define the labels for the pie chart
mylabels = ['<=20', '>20 & <=25', '>25 & <=30', '>30']

# Set the title for the pie chart
plt.title('Total players with age group', fontsize=20)

# Create the pie chart
plt.pie(x, labels=mylabels, autopct="%.1f%%")

# Show the pie chart
plt.show()
```

```python
# Under 20 players in each club
players_u20 = pl_df[pl_df['Age'] < 20]
players_u20['Club'].value_counts().plot(kind='bar', color=sns.color_palette("icefire"))
```

```python
# U20 players in liverpool
players_u20[players_u20['Club'] == 'Manchester City']
```

```python
#Average age of players in each club
plt.figure(figsize=(13,6))
sns.boxplot(x = 'Club', y='Age', data = pl_df)
plt.xticks(rotation=90)
```
